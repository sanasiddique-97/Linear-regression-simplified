```python
import numpy as np # linear algebra
import pandas as pd
```

```python
df = pd.read_csv("CO2 Emissions.csv")
print(df.shape)
df.head(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
sel_columns = [  'Engine Size(L)', 'Cylinders',
        'Fuel Consumption City (L/100 km)',
       'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)',
       'Fuel Consumption Comb (mpg)', 'CO2 Emissions(g/km)']
```

```python
df_sel = df[sel_columns]
df_sel.head(5)
```

Features and label
X: features, the input of the model

y: label, the output of the model, which is the co2 emission amout

```python
# a quick review of slicing and index: usage of : and -1
#list, string
A = [1, 2, 5 ,6]
A[-1]
A[:]
```

```python
X = df_sel.iloc[:,:-1]
y = df_sel.iloc[:,-1]
```

```python
X.isnull()
```

```python
df_sel = df_sel.dropna()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train) # 擬合
model.score(X_test, y_test)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
#VISUALIZATION

import matplotlib.pyplot as plt
plt.figure(figsize = (10,4))
plt.subplot(121)
plt.scatter(y_train, y_pred_train)
plt.plot(y_train, y_train, 'r:')
plt.xlabel('Ground Truth')
plt.ylabel('Prediction')
plt.title('Trianing Data')
plt.grid()
plt.subplot(122)
plt.scatter(y_test, y_pred_test)
plt.plot(y_test, y_test, 'r:')
plt.xlabel('Ground Truth')
plt.ylabel('Prediction')
plt.title('Test Data')

plt.grid()
```
